```
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.losses import MeanSquaredError # Import MeanSquaredError


# ... (rest of the code remains the same) ...
# Step 2: Load dataset
df = pd.read_csv("/content/Housing.csv")  # Replace with actual filename

# Step 3: Convert categorical variables to numerical values
df = df.replace({'yes': 1, 'no': 0, 'furnished': 2, 'semi-furnished': 1, 'unfurnished': 0})

# Step 4: Define features (X) and target (y)
X = df.drop(columns=['price'])
y = df['price']

# Step 5: Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 6: Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Step 7: Build ANN model
model = Sequential([
    Dense(15, activation='tanh', input_shape=(X_train.shape[1],)),
    Dense(20, activation='tanh'),
    Dense(15, activation='tanh'),
    Dense(1, activation='linear')
])

# Step 8: Compile the model
model.compile(optimizer=keras.optimizers.SGD(), loss=MeanSquaredError(), metrics=['mse'])

# Step 9: Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=16, validation_data=(X_test, y_test))

# Step 10: Evaluate the model
train_loss, train_mse = model.evaluate(X_train, y_train)
test_loss, test_mse = model.evaluate(X_test, y_test)
print(f"Training MSE: {train_mse}, Testing MSE: {test_mse}")

# Step 11: Save the trained model
model.save("housing_price_model.h5")

# Step 12: Define function to load and use the saved model
def predict_price(input_data):
    model = keras.models.load_model("/content/housing_price_model.h5")
    input_data_scaled = scaler.transform(np.array(input_data).reshape(1, -1))
    prediction = model.predict(input_data_scaled)
    return prediction[0][0]

# Step 13: Example usage
example_input = X_test[0]  # Replace with actual input
predicted_price = predict_price(example_input)
print(f"Predicted price: {predicted_price}")
```
